```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib
import warnings
warnings.filterwarnings('ignore')
```

### numpy, tensorflow에 다 들어있지만 들어있는 함수가 이렇다는 것 인지
* 데이터 특성에 따라 선택

### step function : 이진분류

```python
def step(x):
    result = x > 0.000001 #True , Fasle
    return result.astype(np.int) #정수로 변환
```

```python
x = np.arange(-10.0, 10.0, 0.1)
y = step(x)
plt.plot(x, y)
plt.show()
```

### sigmoid function(0 ~ 1)

```python
def sigmoid(x):
    return 1.0 / ( 1.0 + np.exp(-x))
```

```python
#딥러닝은 무조건 실수!
x = np.arange(-10.,10., 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.show
```

### tanh function (-1 ~ 1) : 출력값이 음수가 필요한 경우 사용

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-np.pi, np.pi, 60)
y = tanh(x)
plt.plot(x, y)
plt.show()
```

### Relu Function(0보다 작으면 0으로, 크면 그대로) : 음의 값이 필요없는 경우

```python
def relu(x):
    return np.maximum(x, 0)
```

```python
x = np.arange(-10., 10., 0.1)
y = relu(x)
plt.plot(x, y)
plt.show
```
