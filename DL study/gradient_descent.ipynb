## Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = 10
learning_rate = 0.2
max_iter = 50
```

```python
# loss function
loss_func = lambda x: (x-3)**2 + 10
```

```python
# gradient function : 손실함수의 1차 미분
gradient = lambda x: 2*x-6
```

```python
list1 = []
list2 = []
```

```python
# 강하법
for i in range(max_iter): #50번까지 반복
    x = x - learning_rate * gradient(x)
    list1.append(x)
    list2.append(loss_func(x))
    print('손실함수값(', x , ')', loss_func(x))
```

```python
# max_iter, learning_rate가 hyper parameter -> 하나씩 넣어서 확인해봐야함(자동화 함수 필요성)
x1 = np.linspace(0.0, 10.0)
y1 = loss_func(x1)
fig, ax = plt.subplots()
ax.plot(x1, y1)
ax.plot(list1, list2, '*')
```
